```python
# Loading Packages

import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
# Importing data from GitHub

df = pd.read_csv('https://github.com/ArinB/MSBA-CA-Data/raw/main/CA05/movies_recommendation_data.csv')
df
```

### Fitting the kNN Model

```python
# Selecting the relevant features (Rating + Genre Encodings)
model_df = df.iloc[:,2:10]

# Fitting the kNN model
neigh = NearestNeighbors(n_neighbors=5)
neigh.fit(model_df)
```

### Predictions

```python
# Creating the prediction dataframe for "The Post"
prediction_df = pd.DataFrame({'IMDB Rating' : [7.2],
                              'Biography': [1],
                              'Drama': [1],
                              'Thriller': [0],
                              'Comedy': [0],
                              'Crime': [0],
                              'Mystery': [0],
                              'History': [1]})
prediction_df
```

```python
# Finding five of the closest observations in the dataset
neighbors = neigh.kneighbors(prediction_df)
neighbors
```

### Results

```python
# Indices of 5 nearest neighbors in df
arr = np.array(neighbors[1])

# Names of Movie Reccomendations
recs = df['Movie Name'].iloc[arr[0]]
print(recs)
```
